```python
Q1_price = house['Price'].describe()['25%']
Q3_price = house['Price'].describe()['75%']
IQR = Q3_price - Q1_price
out_price = house[(house['Price'] >= Q1_price - (1.5*IQR)) & (house['Price'] <= Q3_price + (1.5 * IQR))]
out_price.groupby('Car').describe()['Price'][['25%','75%','min','max','mean','std', '50%']]
```

```python
Q1_land = house['Landsize'].describe()['25%']
Q3_land = house['Landsize'].describe()['75%']
IQR = Q3_land - Q1_land
out_land = house[(house['Landsize'] >= Q1_land - (1.5*IQR)) & (house['Landsize'] <= Q3_land + (1.5 * IQR))]
out_land.groupby('Car').describe()['Landsize'][['25%','75%','min','max','mean','std', '50%']]
```

```python
land = out_land.groupby('Car').median()['Landsize']
```

```python
def fill_car(x):
    if x['Landsize'] < land[0] and pd.isna(x['Car']) and pd.isna(x['Car']):
        return 0
    elif x['Landsize'] in pd.Interval(land[0],land[1]) and pd.isna(x['Car']):
        return 1
    elif x['Landsize'] in pd.Interval(land[1],land[2]) and pd.isna(x['Car']):
        return 2
    elif x['Landsize'] in pd.Interval(land[2],land[3]) and pd.isna(x['Car']):
        return 3
    elif x['Landsize'] in pd.Interval(land[3],land[4]) and pd.isna(x['Car']):
        return 4
    elif x['Landsize'] in pd.Interval(land[4], (650+699)/2) and pd.isna(x['Car']):
        return 5
    elif x['Landsize'] in pd.Interval((650+699)/2, land[6]) and pd.isna(x['Car']):
        return 6
    elif x['Landsize'] in pd.Interval(land[6], land[7]) and pd.isna(x['Car']):
        return 7
    elif x['Landsize'] in pd.Interval(land[7], land[8]) and pd.isna(x['Car']):
        return 8
    elif x['Landsize'] in pd.Interval(land[8], land[9]) and pd.isna(x['Car']):
        return 9
    elif x['Landsize'] in pd.Interval(land[9], land[10]) or x['Landsize'] > land[10] and pd.isna(x['Car']):
        return 10
    else:
        return x['Car']
    
house['Car'] = house[['Car','Landsize']].apply(fill_car, axis = 1)
```
